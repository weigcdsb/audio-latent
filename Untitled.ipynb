```python
from pathlib import Path

import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt

from scipy.signal import stft
from tqdm.auto import tqdm
```

```python
BIRDS = [
    "Bl122_ChronicLeftAAC",
    "Li145_ChronicLeftAAC",
    "Or61_ChronicLeftAAC",
    "Ti81_ChronicLeftAAC",
]
```

```python
BASE_DIR = Path.cwd() / "yang_long_2025_budgie_aac"
DATA_DIR = BASE_DIR / "data"

BIRD = BIRDS[0]

# same frequency range for analysis and visualization
FREQ_MIN_HZ = 300
FREQ_MAX_HZ = 14500

# same basic STFT settings as before
STFT_WINDOW_SEC = 0.02
STFT_STEP_SEC = 0.001
NFFT = 1024

# search a representative event
SCAN_START_SEC = 0
SCAN_DURATION_SEC = 600
SEARCH_WINDOW_SEC = 10

# event constraints
MIN_EVENT_MS = 30
MAX_EVENT_MS = 700
PAD_MS = 30

# display
DYNAMIC_RANGE_DB = 70
BACKGROUND_CLAMP_FRAC = 0.20

# optional display-only denoising
USE_NOISEREDUCE = False

audio_path = DATA_DIR / BIRD / "audioCh3_HP.flac"
assert audio_path.exists(), audio_path

print(audio_path)
print(sf.info(str(audio_path)))
```

```python
try:
    import noisereduce as nr
    HAS_NOISEREDUCE = True
except ImportError:
    HAS_NOISEREDUCE = False

print("HAS_NOISEREDUCE:", HAS_NOISEREDUCE)
print("USE_NOISEREDUCE:", USE_NOISEREDUCE)

if USE_NOISEREDUCE and not HAS_NOISEREDUCE:
    print("Noisereduce requested but not installed. Install with: pip install noisereduce")
```

```python
def read_audio_chunk(bird, start_sec, duration_sec):
    path = DATA_DIR / bird / "audioCh3_HP.flac"
    info = sf.info(str(path))
    fs = info.samplerate

    start = int(round(start_sec * fs))
    stop = int(round((start_sec + duration_sec) * fs))
    stop = min(stop, info.frames)

    x, fs = sf.read(str(path), start=start, stop=stop, always_2d=False)

    if x.ndim > 1:
        x = x[:, 0]

    return x.astype(np.float32), fs


def denoise_audio(x, fs):
    """
    Optional Noisereduce-style spectral gating.
    Used for visualization only.
    """
    if not USE_NOISEREDUCE:
        return x

    if not HAS_NOISEREDUCE:
        return x

    return nr.reduce_noise(
        y=x,
        sr=fs,
        stationary=False,
        prop_decrease=0.95,
        time_constant_s=1.0,
    ).astype(np.float32)


def compute_sonogram(x, fs):
    """
    STFT amplitude sonogram.
    Use amplitude, not PSD.
    """
    nperseg = int(round(STFT_WINDOW_SEC * fs))
    hop = int(round(STFT_STEP_SEC * fs))
    noverlap = nperseg - hop

    freq, time, Z = stft(
        x,
        fs=fs,
        window="hann",
        nperseg=nperseg,
        noverlap=noverlap,
        nfft=NFFT,
        boundary=None,
        padded=False,
    )

    mask = (freq >= FREQ_MIN_HZ) & (freq <= FREQ_MAX_HZ)

    freq = freq[mask]
    S_amp = np.abs(Z[mask]).T
    S_db = 20 * np.log10(S_amp + 1e-12)

    return S_db, time, freq


def db_to_unit_range(S_db):
    """
    Convert dB sonogram to [0, 1] using fixed dynamic range.
    """
    S_rel = S_db - np.max(S_db)
    S = (S_rel + DYNAMIC_RANGE_DB) / DYNAMIC_RANGE_DB
    return np.clip(S, 0, 1)


def background_clamp(S):
    """
    Visualization only:
    set weak background to zero.
    """
    S_vis = S.copy()
    threshold = S_vis.min() + BACKGROUND_CLAMP_FRAC * (S_vis.max() - S_vis.min())
    S_vis[S_vis < threshold] = 0
    return S_vis
```

```python
def smooth_1d(x, width):
    width = int(max(1, width))
    ker = np.ones(width) / width
    return np.convolve(x, ker, mode="same")


def runs_from_mask(mask):
    idx = np.flatnonzero(mask)

    if len(idx) == 0:
        return []

    runs = []
    s = idx[0]
    p = idx[0]

    for i in idx[1:]:
        if i == p + 1:
            p = i
        else:
            runs.append((s, p + 1))
            s = i
            p = i

    runs.append((s, p + 1))
    return runs


def spectral_flatness(S):
    """
    S: time x freq, nonnegative.
    Smaller flatness = more tonal / harmonic.
    """
    S = S + 1e-8
    return np.exp(np.mean(np.log(S), axis=1)) / np.mean(S, axis=1)


def find_representative_event_in_chunk(x, fs):
    """
    Detect candidate vocal events and select one that is both strong and tonal.
    """
    x_denoised = denoise_audio(x, fs)
    S_db, tt, freq = compute_sonogram(x_denoised, fs)
    S = db_to_unit_range(S_db)

    energy_t = S.mean(axis=1)
    tonal_t = 1 - spectral_flatness(S)

    score_t = smooth_1d(energy_t * tonal_t, width=15)

    thr = np.quantile(score_t, 0.90)
    active = score_t > thr

    runs = runs_from_mask(active)

    candidates = []

    for s, e in runs:
        dur_ms = (e - s) * STFT_STEP_SEC * 1000

        if dur_ms < MIN_EVENT_MS or dur_ms > MAX_EVENT_MS:
            continue

        score = score_t[s:e].mean()

        candidates.append((score, s, e))

    if len(candidates) == 0:
        return None

    candidates = sorted(candidates, reverse=True)
    _, s, e = candidates[0]

    pad = int(round(PAD_MS / 1000 / STFT_STEP_SEC))
    s = max(0, s - pad)
    e = min(len(tt), e + pad)

    start_sec = tt[s]
    end_sec = tt[e - 1]

    return start_sec, end_sec, score_t, S
```

```python
def find_representative_event(bird):
    best = None

    starts = np.arange(
        SCAN_START_SEC,
        SCAN_START_SEC + SCAN_DURATION_SEC,
        SEARCH_WINDOW_SEC,
    )

    for start_sec in tqdm(starts, desc=f"search {bird}"):
        x, fs = read_audio_chunk(bird, start_sec, SEARCH_WINDOW_SEC)
        out = find_representative_event_in_chunk(x, fs)

        if out is None:
            continue

        event_start, event_end, score_t, _ = out

        # score again by local peak
        s = int(round(event_start / STFT_STEP_SEC))
        e = int(round(event_end / STFT_STEP_SEC))
        score = score_t[max(0, s):min(len(score_t), e)].max()

        abs_start = start_sec + event_start
        abs_end = start_sec + event_end

        if best is None or score > best["score"]:
            best = {
                "bird": bird,
                "start_sec": float(abs_start),
                "end_sec": float(abs_end),
                "duration_ms": float((abs_end - abs_start) * 1000),
                "score": float(score),
            }

    return best


best_event = find_representative_event(BIRD)
best_event
```

```python
def plot_representative_event(best_event):
    bird = best_event["bird"]
    start_sec = best_event["start_sec"]
    end_sec = best_event["end_sec"]

    duration_sec = end_sec - start_sec

    x_raw, fs = read_audio_chunk(bird, start_sec, duration_sec)
    x_vis = denoise_audio(x_raw, fs)

    S_db_raw, tt, freq = compute_sonogram(x_raw, fs)
    S_db_vis, _, _ = compute_sonogram(x_vis, fs)

    S_analysis = db_to_unit_range(S_db_raw)
    S_display = background_clamp(db_to_unit_range(S_db_vis))

    t_ms = tt * 1000
    extent = [t_ms[0], t_ms[-1], freq[0] / 1000, freq[-1] / 1000]

    fig, axes = plt.subplots(
        2, 1,
        figsize=(7, 7),
        sharex=True,
        sharey=True,
    )

    axes[0].imshow(
        S_analysis.T,
        origin="lower",
        aspect="auto",
        interpolation="nearest",
        cmap="turbo",
        vmin=0,
        vmax=1,
        extent=extent,
    )
    axes[0].set_title(
        "analysis sonogram: raw STFT amplitude, no background clamp"
    )
    axes[0].set_ylabel("frequency (kHz)")

    axes[1].imshow(
        S_display.T,
        origin="lower",
        aspect="auto",
        interpolation="nearest",
        cmap="turbo",
        vmin=0,
        vmax=1,
        extent=extent,
    )

    denoise_text = "Noisereduce + " if USE_NOISEREDUCE and HAS_NOISEREDUCE else ""
    axes[1].set_title(
        f"visualization sonogram: {denoise_text}{int(BACKGROUND_CLAMP_FRAC * 100)}% background clamp"
    )
    axes[1].set_ylabel("frequency (kHz)")
    axes[1].set_xlabel("time from event start (ms)")

    fig.suptitle(
        f"{bird}\n"
        f"{start_sec:.2f}–{end_sec:.2f} sec, "
        f"{best_event['duration_ms']:.0f} ms, "
        f"{FREQ_MIN_HZ/1000:.1f}–{FREQ_MAX_HZ/1000:.1f} kHz",
        y=1.02,
    )

    plt.tight_layout()
    plt.show()


plot_representative_event(best_event)
```

```python
manual_event = {
    "bird": BIRD,
    "start_sec": 102.0,   # change manually
    "end_sec": 104.0,    # change manually
    "duration_ms": 350,
    "score": np.nan,
}

plot_representative_event(manual_event)
```
